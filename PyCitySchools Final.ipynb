# District Summary

```python
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
all_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
all_data.head()
```

### District Summary

```python
# Total number of unique schools
school_count = all_data['school_name'].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = len(all_data['student_name'])
student_count
```

```python
# Create a data frame with the budget amount for each school
school_budget = all_data.groupby(['school_name', 'budget']).first()
school_budget
```

```python
school_budget =school_budget.reset_index()
```

```python
school_budget
```

```python
budget_amount = school_budget['budget'].sum()
budget_amount
```

```python
# Calculate the average (mean) math score
average_math_score = all_data['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = all_data['reading_score'].mean()
average_reading_score
```

```python
# Percentage of students Passing MATH
passing_math_count = all_data[(all_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Percentage of students Passing READING
passing_reading_count = all_data[(all_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Percentage of students passing BOTH MATH AND READING
passing_math_reading_count = all_data[
    (all_data["math_score"] >= 70) & (all_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

# District Summary Numbers

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Number of Unique Schools": school_count,
    "Total Students": student_count,
    "Total Budget": budget_amount,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing": overall_passing_rate
}, index=[0])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending = per_school_summary.copy()
school_spending
```

# Scores by School Spending


```python
# Use `pd.cut` to categorize spending based on the bins.
#Had to take the $ away to get this to work
school_spending['Per Capita'] = pd.to_numeric(
    school_spending['Per Capita'].str.replace('$', ''),
    errors='coerce'
)

school_spending['Spending Ranges (Per Student)'] = pd.cut(
    school_spending['Per Capita'],
    bins=spending_bins,
    labels=labels,
    include_lowest=True
)

school_spending
```

```python
#  Calculate averages for the desired columns. 

#was getting an error about % Passing Math not having numeric values
school_spending["% Passing Math"] = pd.to_numeric(school_spending["% Passing Math"])
school_spending["% Passing Math"] = pd.to_numeric(school_spending["% Passing Math"], errors="coerce")
school_spending["% Passing Reading"] = pd.to_numeric(school_spending["% Passing Reading"], errors="coerce")
school_spending["% Overall Passing"] = pd.to_numeric(school_spending["% Overall Passing"], errors="coerce")
school_spending["Average Math Score"] = pd.to_numeric(school_spending["Average Math Score"], errors="coerce")
school_spending["Average Reading Score"] = pd.to_numeric(school_spending["Average Reading Score"], errors="coerce")



spending_math_scores = school_spending.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
```
